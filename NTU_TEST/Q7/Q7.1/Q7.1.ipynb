```python
def coordinate_to_index(x1,x2):
    if 0<= x1 <= 49 and 0<= x2 <= 56:
        Index = x1 + x2*50
        print(x1,x2,Index)
        return Index
    else:
        print('The input coordinate is out of bound')
    
```

```python
def index_to_coordinate(index):
    if index >=0 and index <= 2849:
        x2 = index// 50
        x1 = index - x2 * 50
        print(x1,x2)
        return x1,x2
```
